```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = results_da.plot_predict(dynamic=False, ax=ax);
```

```python
n_dam_scenarios = 700
pi = 1/n_dam_scenarios
prob = [pi for i in range(n_dam_scenarios)]
ar = np.r_[1, -results_da.arparams]  # 1 is coefficient on the zero-lag; (-) sign is convention
ma = np.r_[1, results_da.maparams]
c = results_da.params[0]
y_all = pd.DataFrame()
for s in range(n_dam_scenarios):
    np.random.seed(s)
    y = c+pd.Series(arma_generate_sample(ar, ma, nsample=len(prices), sigma=np.sqrt(results_da.sigma2), burnin=0))
    y_all = pd.concat([pd.DataFrame(y), y_all], axis=1)
```

```python
balPos_scen.rename(columns={0: "ImbPos1", 1: "ImbPos2", 2: "ImbPos3"}, inplace=True)
balNeg_scen.rename(columns={0: "ImbNeg1", 1: "ImbNeg2", 2: "ImbNeg3"}, inplace=True)
```

```python
def prepare_scenarios(dam_scen, pv_scen, balPos_scen, balNeg_scen, n_dam, n_pv, n_imb):
    
    pPriceDA = np.vstack(((np.array(dam_scen['DAM1']), np.array(dam_scen['DAM2']), np.array(dam_scen['DAM3']))
                         for i in range(n_pv*n_imb))).T
    pPV = np.vstack(([np.array(pv_scen['PV1']) for j in range(n_dam*n_imb)], [np.array(pv_scen['PV2']) for j in range(n_dam*n_imb)],
                    [np.array(pv_scen['PV3']) for j in range(n_dam*n_imb)])).T
    
    pPosImb = np.vstack(((np.array(balPos_scen['ImbPos1']), np.array(balPos_scen['ImbPos2']), np.array(balPos_scen['ImbPos3']))
                         for i in range(n_pv))).T
    pNegImb = np.vstack(((np.array(balNeg_scen['ImbNeg1']), np.array(balNeg_scen['ImbNeg2']), np.array(balNeg_scen['ImbNeg3']))
                         for i in range(n_pv))).T
#     pPosImb = np.vstack(([np.array(balPos_scen['ImbPos1']) for j in range(n_dam*n_pv)], [np.array(balPos_scen['ImbPos2']) for j in range(n_dam*n_pv)],
#                     [np.array(balPos_scen['ImbPos3']) for j in range(n_dam*n_pv)])).T
#     pNegImb = np.vstack(([np.array(balNeg_scen['ImbNeg1']) for j in range(n_dam*n_pv)], [np.array(balNeg_scen['ImbNeg2']) for j in range(n_dam*n_pv)],
#                     [np.array(balNeg_scen['ImbNeg3']) for j in range(n_dam*n_pv)])).T
    return pPriceDA, pPV, pPosImb, pNegImb
```

```python
n_dam = 3 # of DAM price scenarios
n_pv = 3  # of PV production scenarios
n_imb = 1 # of Imbalance scenarios (positive and negative) per DAM price scenario
pPriceDA, pPV, pPosImb, pNegImb = prepare_scenarios(dam_scen, pv_scen, balPos_scen, balNeg_scen, n_dam, n_pv, n_imb)
```
